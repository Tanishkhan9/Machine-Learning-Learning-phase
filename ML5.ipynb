Core Steps Covered
Data Loading:
Reads a dataset (CSV file) containing student study hours and their test scores using pandas.

Data Preparation:
Splits the data into a feature (Hours—input) and a target (Scores—output).

Model Training:
Uses scikit-learn’s LinearRegression to train a simple predictive model that fits a line to the data (y = mx + b).

Prediction:
The trained model predicts test scores for the same set of study hours.

Performance Evaluation:
Measures how good the predictions are using three common regression metrics:

Mean Absolute Error (MAE): Average absolute difference between predicted and actual scores.
Mean Squared Error (MSE): Average squared difference, penalizing larger errors more.
Root Mean Squared Error (RMSE): Square root of MSE; error in the original score units.
Result Output:
Prints out the error metrics so you can easily see how accurate the model is.

Key Learnings
Shows the basic workflow of a supervised machine learning problem: data loading, preprocessing, model training, prediction, and evaluation.
Introduces essential tools in Python’s data science ecosystem: pandas, scikit-learn, and NumPy.
Emphasizes the importance of evaluating model performance using understandable numerical metrics.


```python
# Import regression evaluation metrics from scikit-learn
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Import pandas for data manipulation and CSV file handling
import pandas as pd

# Import Linear Regression algorithm from scikit-learn
from sklearn.linear_model import LinearRegression

# Import NumPy for numerical operations (specifically for square root calculation)
import numpy as np

# Load the student dataset from CSV file
# Expected columns: 'Hours' (study time) and 'Scores' (test results)
df = pd.read_csv('/content/student.csv')

# Prepare the feature matrix (independent variable)
# Double brackets [[]] create a 2D DataFrame required by sklearn
X = df[['Hours']]  # Study hours - input feature

# Prepare the target vector (dependent variable)
# Single brackets [] create a 1D Series which is acceptable for target
y = df['Scores']   # Test scores - what we want to predict

# Create a Linear Regression model instance
model = LinearRegression()

# Train the model on the dataset
# The model learns the relationship: Scores = slope * Hours + intercept
model.fit(X, y)

# Generate predictions using the trained model
# Using the same data for training and testing (not ideal in real scenarios)
Prediction = model.predict(X)

# Calculate Mean Absolute Error
# Formula: Average of |actual - predicted| values
# Measures average prediction error in same units as target variable
mae = mean_absolute_error(y, Prediction)

# Calculate Mean Squared Error
# Formula: Average of (actual - predicted)² values
# Penalizes larger errors more heavily due to squaring
mse = mean_squared_error(y, Prediction)

# Calculate Root Mean Squared Error
# Formula: Square root of MSE
# Brings error metric back to original units, making it more interpretable
rmse = np.sqrt(mse)

# Display all regression evaluation metrics
print('MAE:', mae)    # Lower values indicate better model performance
print('MSE:', mse)    # Lower values indicate better model performance
print('RMSE:', rmse)  # Lower values indicate better model performance
```

This Python script demonstrates a complete, basic machine learning workflow using a simple linear regression model. The primary goal is to predict a student's academic score based on the number of hours they study. The script uses popular libraries like pandas, scikit-learn, and matplotlib to handle data, build the model, and visualize the results.

The overall process is broken down into the following key steps:

Data Preparation:

The script loads a student dataset from a student.csv file into a pandas DataFrame.
It separates the data into the feature (input variable X), which is 'Hours', and the target (output variable y), which is 'Scores'.
Model Training:

An instance of the LinearRegression model from scikit-learn is created.
The model is trained using the .fit() method on the 'Hours' and 'Scores' data. During this step, the model learns the best-fit line that describes the relationship between the two variables.
Model Evaluation:

The trained model's performance is assessed by comparing its predictions to the actual scores.
Four standard regression metrics are calculated and printed:
Mean Absolute Error (MAE): The average absolute difference between predicted and actual scores.
Mean Squared Error (MSE): The average of the squared differences, which penalizes larger errors.
Root Mean Squared Error (RMSE): The square root of MSE, providing an error metric in the same units as the target.
R-squared (R²): A value that indicates the proportion of the variance in scores that can be explained by study hours.
Visualization:

The script generates two plots to analyze the data and the model's performance:
Histogram: Shows the distribution of student scores to understand their frequency and spread.
Scatter Plot with Regression Line: Plots the actual data points and overlays the model's regression line. This provides a clear visual representation of how well the model fits the data.
Prediction on New Data:

Finally, the script demonstrates the practical use of the trained model by predicting a score for a new, unseen input (e.g., a student who studies for 9 hours).

```python
# Import necessary libraries
import pandas as pd                    # For data handling and CSV file reading
import matplotlib.pyplot as plt        # For creating visualizations
import numpy as np                     # For numerical operations (like sqrt)
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression # The machine learning model

# Load the student performance dataset from a CSV file
data = pd.read_csv('/content/student.csv')

# --- 1. Data Preparation ---
# Define the feature (X) and target (y)
X = data[['Hours']]  # Independent variable (input). Double brackets make it a 2D DataFrame.
y = data['Scores']   # Dependent variable (output).

# --- 2. Model Training ---
# Create an instance of the Linear Regression model
model = LinearRegression()

# Train the model with our data (X and y)
# The model learns the best-fit line: Scores = slope * Hours + intercept
model.fit(X, y)

# --- 3. Making Predictions ---
# Use the trained model to predict scores based on the hours in our dataset
Predicted_score = model.predict(X)

# --- 4. Model Evaluation ---
# Calculate regression metrics to see how well the model performed

# Mean Absolute Error: The average absolute difference between actual and predicted scores.
mae = mean_absolute_error(y, Predicted_score)

# Mean Squared Error: The average of the squared differences. Penalizes larger errors more.
mse = mean_squared_error(y, Predicted_score)

# Root Mean Squared Error: The square root of MSE. Interpretable in the same units as the target.
rmse = np.sqrt(mse)

# R-squared (R²): The proportion of the variance in the scores that is predictable from the hours.
# A value of 1.0 is a perfect fit.
r2 = r2_score(y, Predicted_score)

# Print the rounded evaluation metrics for readability
print("MAE: ", round(mae, 2))
print("MSE: ", round(mse, 2))
print("RMSE: ", round(rmse, 2))
print("R2 Score : ", round(r2, 4))


# --- 5. Visualization ---
# Plot 1: Histogram to show the distribution of student scores
plt.figure(figsize=(10, 5))
# This plot shows how many students achieved scores within certain ranges.
plt.hist(data['Scores'], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Student Scores") # Corrected Title
plt.xlabel("Scores")                        # Corrected X-axis label
plt.ylabel("Frequency")                     # Corrected Y-axis label
plt.grid(True)
plt.show()

# Plot 2: Scatter plot of actual data vs. the model's regression line
plt.figure(figsize=(10, 6))
# Plot the original data points
plt.scatter(X, y, color='blue', label='Actual Score')
# Plot the line that our model learned
plt.plot(X, Predicted_score, color="red", linewidth=3, label="Predicted score (Regression Line)")
plt.title("Model Prediction vs Actual Score")
plt.xlabel("Study Hours")  # Corrected X-axis label
plt.ylabel("Scores")       # Corrected Y-axis label
plt.legend()               # Show the labels for the plots
plt.grid(True)
plt.show()

# --- 6. Predicting a New Value ---
# Define a new input value for which we want a prediction
new_hours = 9
# Use the trained model to predict the score for the new input
# The input must be a 2D array, so we use [[new_hours]]
predicted_new_score = model.predict([[new_hours]])

# Print the prediction in a user-friendly format
# [0] is used to extract the single value from the prediction array
print(f"Predicted score for studying {new_hours} hours is: {predicted_new_score[0]:.2f}")
```
