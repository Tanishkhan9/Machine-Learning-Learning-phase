to draw line , circle , rectangle

```python
import cv2  # Import OpenCV

image = cv2.imread("F:/project/tests/al.jpg")  # Load image

if image is not None:  # Check if image loaded
    print("Image is loaded successfully")
else:
    print("error")

line = cv2.line(image, (100, 100), (100, 299), (255, 0, 0), 2)  # Draw blue line

cv2.imshow("Original Image", image)  # Show image
cv2.waitKey(0)  # Wait for key press
cv2.destroyAllWindows()  # Close all windows
```

to put text on image


```python
import cv2  # Import OpenCV

image = cv2.imread("F:/project/tests/al.jpg")  # Load image

if image is not None:  # Check if image loaded
    print("Image is loaded successfully")
else:
    print("Error: Image not found")

cv2.putText(image, "Opencv", (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 2.0, (255, 255, 255), 4)  # Add white text

cv2.imshow("image with text", image)  # Show image
cv2.waitKey(0)  # Wait for key press
cv2.destroyAllWindows()  # Close all windows
```
