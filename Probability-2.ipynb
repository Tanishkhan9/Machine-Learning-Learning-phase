# Bayes' Theorem

Bayes' Theorem is a mathematical formula used to determine the conditional probability of an event based on prior knowledge and new evidence.

It adjusts probabilities when new information comes in and helps make better decisions in uncertain situations.

# Bayes Theorem and Conditional Probability

Bayes' theorem (also known as the Bayes Rule or Bayes Law) is used to determine the conditional probability of event A when event B has already occurred.

The general statement of Bayes’ theorem is “The conditional probability of an event A, given the occurrence of another event B, is equal to the product of the probability of B, given A, and the probability of A divided by the probability of event B.” i.e.


For example, if we want to find the probability that a white marble drawn at random came from the first bag, given that a white marble has already been drawn, and there are three bags each containing some white and black marbles, then we can use Bayes’ Theorem.

# Bayes Theorem Formula

For any two events A and B, Bayes's formula for the Bayes theorem is given by:



$$
P(A|B) = \frac{P(B|A) \, P(A)}{P(B)}
$$

Where:

- \(P(A)\) and \(P(B)\) are the probabilities of events \(A\) and \(B\); also, \(P(B) \neq 0\).  
- \(P(A|B)\) is the probability of event \(A\) given that event \(B\) occurs.  
- \(P(B|A)\) is the probability of event \(B\) given that event \(A\) occurs.  


# Bayes Theorem Statement

Bayes' Theorem for n sets of events is defined as,

Let E1, E2,…, En be a set of events associated with the sample space S, in which all the events E1, E2,…, En have a non-zero probability of occurrence. All the events E1, E2,…, E form a partition of S. Let A be an event in space S for which we have to find the probability, then according to Bayes theorem,

# Extended Bayes' Theorem

The formula is:

$$
P(E_i \mid A) = \frac{P(E_i) \cdot P(A \mid E_i)}{\sum_{k=1}^{n} P(E_k) \cdot P(A \mid E_k)},
\quad \text{for } k = 1, 2, 3, \ldots, n
$$


# Bayes Theorem Derivation

# Proof of Bayes' Theorem

According to the conditional probability formula:

$$
P(E_i \mid A) = \frac{P(E_i \cap A)}{P(A)} \tag{i}
$$

---

By the multiplication rule of probability:

$$
P(E_i \cap A) = P(E_i) \cdot P(A \mid E_i) \tag{ii}
$$

---

Now, by the **total probability theorem**:

$$
P(A) = \sum_{k=1}^{n} P(E_k) \cdot P(A \mid E_k) \tag{iii}
$$

---

Substituting the values of \(P(E_i \cap A)\) from (ii) and \(P(A)\) from (iii) into (i), we get:

$$
P(E_i \mid A) = \frac{P(E_i) \cdot P(A \mid E_i)}{\sum_{k=1}^{n} P(E_k) \cdot P(A \mid E_k)}
$$



Bayes’ theorem is also known as the formula for the probability of “causes”. As we know, the Ei‘s are a partition of the sample space S, and at any given time, only one of the events Ei occurs. Thus, we conclude that the Bayes theorem formula gives the probability of a particular Ei, given that event A has occurred.


# Terms Related to Bayes' Theorem

After learning about Bayes theorem in detail, let us understand some important terms related to the concepts we covered in the formula and derivation.

# Key Concepts in Probability and Bayes' Theorem

### Hypotheses
Hypotheses refer to possible events or outcomes in the sample space;  
they are denoted as \$(E1, E2, \ldots , En\)$.

- Each hypothesis represents a distinct scenario that could explain an observed event.

---

### Priori Probability
- **Prior probability** \(P(Ei)\) is the initial probability of an event occurring before any new data is considered.  
- It reflects existing knowledge or assumptions about the event.  
- **Example:** The probability of a person having a disease *before* taking a test.  

---

### Posterior Probability
- **Posterior probability** \(P(Ei \ A)\) is the updated probability of an event after considering new information.  
- It is derived using **Bayes' Theorem**.  
- **Example:** The probability of having a disease *given* a positive test result.  

---

### Conditional Probability
- The probability of an event \(A\) based on the occurrence of another event \(B\) is called **conditional probability**.  
- It is denoted as \(P(A \ B)\), representing the probability of \(A\) when event \(B\) has already occurred.  

---

### Joint Probability
- **Joint probability** measures the probability of two or more events occurring together.  
- For two events \(A\) and \(B\), it is denoted as:  
  $$
  P(A \cap B)
  $$

---

### Random Variables
- **Random variables** are real-valued variables whose possible values are determined by random experiments.  
- The probability associated with such variables is known as the **experimental probability**.

| Aspect | Bayes' Theorem | Conditional Probability |

# Theorem of Total Probability

Let E1, E2,…., En be mutually exclusive and exhaustive events of a sample space S, and let E be any event that occurs with some Ei​. Then, prove that :

# Law of Total Probability

The law of total probability states that if The events are denoted as:

The events are denoted as $E_1, E_2, \ldots, E_n$.


 are mutually exclusive and exhaustive events, then the probability of an event \(E\) can be expressed as:

$$
P(E) = \sum_{i=1}^{n} P(E \mid E_i) \cdot P(E_i)
$$


# Proof of the Law of Total Probability

Let $S$ be the sample space.  
Since the events $E_1, E_2, \ldots, E_n$ are **mutually exclusive** and **exhaustive**, we have:

$$
S = E_1 \cup E_2 \cup E_3 \cup \cdots \cup E_n,
\quad \text{and} \quad E_i \cap E_j = \varnothing \;\; \text{for } i \neq j
$$

---

Now, consider the event $E$:

$$
E = E \cap S
$$

Substituting $S$ with the union of $E_i$’s:

$$
E = E \cap (E_1 \cup E_2 \cup \cdots \cup E_n)
$$

Using the distributive law:

$$
E = (E \cap E_1) \cup (E \cap E_2) \cup \cdots \cup (E \cap E_n)
$$

Since the events $E_i$ are mutually exclusive, the intersections $E \cap E_i$ are also mutually exclusive. Therefore:

$$
P(E) = P\big((E \cap E_1) \cup (E \cap E_2) \cup \cdots \cup (E \cap E_n)\big)
$$

$$
\Rightarrow P(E) = P(E \cap E_1) + P(E \cap E_2) + \cdots + P(E \cap E_n)
$$

(Here, $(E \cap E_1), (E \cap E_2), \ldots, (E \cap E_n)$ are pairwise disjoint.)

---

By the **multiplication rule of probability**:

$$
P(E \cap E_i) = P(E \mid E_i) \cdot P(E_i)
$$

Thus,

$$
P(E) = P(E \mid E_1) \cdot P(E_1) + P(E \mid E_2) \cdot P(E_2) + \cdots + P(E \mid E_n) \cdot P(E_n)
$$

Finally, in summation form:

$$
P(E) = \sum_{i=1}^{n} P(E \mid E_i) \cdot P(E_i)
$$


# Solved Examples of Bayes' Theorem
Example 1: A person has undertaken a job. The probability of completing the job on time if it rains is 0.44, and the probability of completing the job on time if it does not rain is 0.95. If the probability that it will rain is 0.45, then determine the probability that the job will be completed on time.

# Example: Law of Total Probability

Let:  

- $R$: event that it rains  
- $R^c$: event that it does not rain  
- $C$: event that the job is completed on time  

We are given:  

- $P(R) = 0.45$, $P(R^c) = 1 - 0.45 = 0.55$  
- $P(C \mid R) = 0.44$, $P(C \mid R^c) = 0.95$  

---

By the **law of total probability**:  

$$
P(C) = P(R) \cdot P(C \mid R) + P(R^c) \cdot P(C \mid R^c)
$$

Substitute the values:  

$$
P(C) = (0.45)(0.44) + (0.55)(0.95)
$$

$$
P(C) = 0.198 + 0.5225 = 0.7205
$$

---

**Final Answer:**  
The probability that the job is completed on time is:

$$
P(C) = 0.7205 \; \approx \; 72.05\%
$$


**Example 2:** There are three urns containing 3 white and 2 black balls, 2 white and 3 black balls, and 1 black and 4 white balls, respectively.

There is an equal probability of each urn being chosen.

One ball is equal probability chosen at random. What is the probability that a white ball will be drawn?

**Solution:**

# Example: Law of Total Probability (Urn Problem)

Let $E_1, E_2,$ and $E_3$ be the events of choosing the first, second, and third urn respectively.  

$$
P(E_1) = P(E_2) = P(E_3) = \tfrac{1}{3}
$$

---

Let $E$ be the event that a white ball is drawn.  

$$
P(E \mid E_1) = \tfrac{3}{5}, \quad
P(E \mid E_2) = \tfrac{2}{5}, \quad
P(E \mid E_3) = \tfrac{4}{5}
$$

---

By the **theorem of total probability**,  

$$
P(E) = P(E \mid E_1) \cdot P(E_1) \;+\; P(E \mid E_2) \cdot P(E_2) \;+\; P(E \mid E_3) \cdot P(E_3)
$$

Substituting values:  

$$
P(E) = \left(\tfrac{3}{5} \times \tfrac{1}{3}\right)
+ \left(\tfrac{2}{5} \times \tfrac{1}{3}\right)
+ \left(\tfrac{4}{5} \times \tfrac{1}{3}\right)
$$

$$
P(E) = \tfrac{9}{15} = \tfrac{3}{5}
$$

---

**Final Answer:**  
The probability of drawing a white ball is:

$$
P(E) = \tfrac{3}{5} = 0.6
$$


**Example 3:** A card from a pack of 52 cards is lost. From the remaining cards of the pack, two cards are drawn and are found to be both hearts. Find the probability of the lost card being a heart.


**Solution:**

# Example: Bayes' Theorem with Playing Cards

Let $E_1, E_2, E_3,$ and $E_4$ be the events of losing a card of **hearts, clubs, spades,** and **diamonds** respectively.  

Since one card is missing at random:  

$$
P(E_1) = P(E_2) = P(E_3) = P(E_4) = \frac{13}{52} = \frac{1}{4}
$$

---

Let $E$ be the event of drawing **2 hearts** from the remaining 51 cards.  

### Case 1: A heart is missing  
$$
P(E \mid E_1) = \frac{\binom{12}{2}}{\binom{51}{2}}
= \frac{12 \times 11}{51 \times 50} = \frac{22}{425}
$$

### Case 2: A club is missing  
$$
P(E \mid E_2) = \frac{\binom{13}{2}}{\binom{51}{2}}
= \frac{13 \times 12}{51 \times 50} = \frac{26}{425}
$$

### Case 3: A spade is missing  
$$
P(E \mid E_3) = \frac{\binom{13}{2}}{\binom{51}{2}}
= \frac{26}{425}
$$

### Case 4: A diamond is missing  
$$
P(E \mid E_4) = \frac{\binom{13}{2}}{\binom{51}{2}}
= \frac{26}{425}
$$

---

Now, by **Bayes' Theorem**:  

$$
P(E_1 \mid E) = \frac{P(E_1) \cdot P(E \mid E_1)}{P(E_1) \cdot P(E \mid E_1) + P(E_2) \cdot P(E \mid E_2) + P(E_3) \cdot P(E \mid E_3) + P(E_4) \cdot P(E \mid E_4)}
$$

Substitute the values:  

$$
P(E_1 \mid E) = \frac{\tfrac{1}{4} \cdot \tfrac{22}{425}}
{\tfrac{1}{4} \cdot \tfrac{22}{425} + \tfrac{1}{4} \cdot \tfrac{26}{425} + \tfrac{1}{4} \cdot \tfrac{26}{425} + \tfrac{1}{4} \cdot \tfrac{26}{425}}
$$

Simplify:  

$$
P(E_1 \mid E) = \frac{22}{100} = 0.22
$$

---

**Final Answer:**  
The probability that the lost card is a **heart**, given that 2 hearts were drawn from the remaining deck, is:  

$$
P(E_1 \mid E) = 0.22 \; \text{or} \; 22\%
$$

**Example 5: **A man is known to speak the lies 1 out of 4 times. He throws a die and reports that it is a six. Find the probability that it is actually a six.

**Solution: **

# Example: Bayes' Theorem (Witness Problem)

In a throw of a die, let:  

- $E_1 =$ event of getting a six  
- $E_2 =$ event of not getting a six  
- $E =$ event that the man reports that it is a six  

---

### Step 1: Prior probabilities
$$
P(E_1) = \frac{1}{6}, \quad P(E_2) = 1 - \frac{1}{6} = \frac{5}{6}
$$

---

### Step 2: Likelihoods
- If six has actually occurred ($E_1$), the man tells the truth with probability $\tfrac{3}{4}$:

$$
P(E \mid E_1) = \tfrac{3}{4}
$$

- If six has not occurred ($E_2$), the man lies with probability $\tfrac{1}{4}$:

$$
P(E \mid E_2) = \tfrac{1}{4}
$$

---

### Step 3: Bayes' Theorem
We need the probability that a six actually occurred, given the man reported a six:

$$
P(E_1 \mid E) = \frac{P(E \mid E_1) \cdot P(E_1)}{P(E \mid E_1) \cdot P(E_1) + P(E \mid E_2) \cdot P(E_2)}
$$

Substitute the values:

$$
P(E_1 \mid E) = \frac{\tfrac{3}{4} \cdot \tfrac{1}{6}}{\tfrac{3}{4} \cdot \tfrac{1}{6} + \tfrac{1}{4} \cdot \tfrac{5}{6}}
$$

---

### Step 4: Simplification
$$
P(E_1 \mid E) = \frac{\tfrac{3}{24}}{\tfrac{3}{24} + \tfrac{5}{24}} = \frac{3}{8}
$$

---

### ✅ Final Answer
The probability that a six was actually thrown, given the man reported it, is:

$$
P(E_1 \mid E) = \tfrac{3}{8} = 0.375
$$
