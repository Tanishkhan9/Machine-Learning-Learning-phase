```python
import cv2  # Import OpenCV

image = cv2.imread(r"F:\project\tests\BLACK AND WHITE.jpg", cv2.IMREAD_GRAYSCALE)  # Load as grayscale

_, thresh = cv2.threshold(image, 200, 255, cv2.THRESH_BINARY)  # Apply binary threshold

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)  # Find contours

image_color = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)  # Convert gray to BGR for colored contours

cv2.drawContours(image_color, contours, -1, (0, 255, 0), 3)  # Draw contours

cv2.imshow("Contours", image_color)  # Show result
cv2.waitKey(0)  # Wait for keypress
cv2.destroyAllWindows()  # Close windows
```

```python
import cv2  # Import OpenCV

image = cv2.imread(r"F:\project\tests\Rectangle.jpg", cv2.IMREAD_GRAYSCALE)  # Load grayscale image

_, thresh = cv2.threshold(image, 240, 255, cv2.THRESH_BINARY)  # Apply binary threshold

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)  # Find contours

for contour in contours:  # Loop through all contours
    approx = cv2.approxPolyDP(contour, 0.01 * cv2.arcLength(contour, True), True)  # Approximates contour shape

    if len(approx) == 3:
        shape_name = "Triangle Detected"
    elif len(approx) == 4:
        shape_name = "Rectangle Detected"
    elif len(approx) == 5:
        shape_name = "Pentagon Detected"
    else:
        print("Circle Detected")

cv2.drawContours(image, [approx], 0, (0, 255, 0), 2)  # Draw the last detected shape contour

cv2.putText(image, shape_name, (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 255), 3)  # Draw text

cv2.drawContours(image, contours, -1, (0, 255, 0), 3)  # Draw all contours

cv2.imshow("Contours", image)  # Show the result
cv2.waitKey(0)  # Wait for key press
cv2.destroyAllWindows()  # Close all OpenCV windows
```
